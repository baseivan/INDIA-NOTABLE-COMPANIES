```python
# Loaded variable 'df' from URI: c:\Users\HP\Desktop\python data\INDIA NOTABLE COMPANIES\notable_companies (1).csv
import pandas as pd
df = pd.read_csv(r'c:\Users\HP\Desktop\python data\INDIA NOTABLE COMPANIES\notable_companies (1).csv')
# Install necessary packages
%pip install pandas matplotlib seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'c:\Users\HP\Desktop\python data\INDIA NOTABLE COMPANIES\notable_companies (1).csv')

# Clean the data
def clean_data(df):
    # Drop rows with missing data in column: 'Sector'
    df = df.dropna(subset=['Sector'])
    # Fill missing values in 'Notes' column with an empty string
    df['Notes'] = df['Notes'].fillna('')
    # Convert 'Founded' column to datetime
    df['Founded'] = pd.to_datetime(df['Founded'], errors='coerce')
    return df

df_clean = clean_data(df.copy())

# Display basic information about the dataset
print(df_clean.info())

# Display summary statistics
print(df_clean.describe(include='all'))

# Plot the distribution of industries
plt.figure(figsize=(10, 8))
industry_counts = df_clean['Industry'].value_counts()
sns.barplot(y=industry_counts.index, x=industry_counts.values, palette='viridis')
plt.title('Distribution of Industries')
plt.xlabel('Number of Companies')
plt.ylabel('Industry')
plt.show()

# Plot the distribution of headquarters
plt.figure(figsize=(10, 8))
headquarters_counts = df_clean['Headquarters'].value_counts().head(20)  # Top 20 headquarters
sns.barplot(y=headquarters_counts.index, x=headquarters_counts.values, palette='viridis')
plt.title('Top 20 Headquarters by Number of Companies')
plt.xlabel('Number of Companies')
plt.ylabel('Headquarters')
plt.show()

# Plot the distribution of sectors
plt.figure(figsize=(14, 12))
sector_counts = df_clean['Sector'].value_counts()
sns.barplot(y=sector_counts.index, x=sector_counts.values, palette='viridis')
plt.title('Distribution of Sectors')
plt.xlabel('Number of Companies')
plt.ylabel('Sector')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust the layout to make room for the labels
plt.show()

# Plot the distribution of companies by founded year
plt.figure(figsize=(12, 8))
founded_counts = df_clean['Founded'].dt.year.value_counts().sort_index()
sns.lineplot(x=founded_counts.index, y=founded_counts.values, marker='o')
plt.title('Number of Companies Founded by Year')
plt.xlabel('Year')
plt.ylabel('Number of Companies')
plt.grid(True)
plt.show()

# Plot the distribution of companies by Private/State
plt.figure(figsize=(8, 8))
private_state_counts = df_clean['Private/State'].value_counts()
plt.pie(private_state_counts.values, labels=private_state_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(private_state_counts)))
plt.title('Distribution of Companies by Private/State')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()

# Plot the distribution of companies by Active/Defunct
plt.figure(figsize=(8, 8))
active_defunct_counts = df_clean['Active/Defunct'].value_counts()
plt.pie(active_defunct_counts.values, labels=active_defunct_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(active_defunct_counts)))
plt.title('Distribution of Companies by Active/Defunct')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
